## Parte 3 — Diccionario (Cantidad de acciones)

```python
acciones_portafolio = {
    "Apple": 10,
    "Microsoft": 5,
    "Amazon": 8,
    "Tesla": 4,
    "Google": 6
}
print(acciones_portafolio)
```

## Parte 4 — Valor invertido

```python
valor_apple = precios_acciones[0] * acciones_portafolio["Apple"]
print(valor_apple)
```

### Ejercicio
Calcula el valor invertido en Microsoft, Amazon, Tesla y Google.

```python
# Escribe tu código aquí
valor_microsoft = precios_acciones[1] * acciones_portafolio["Microsoft"]
valor_amazon = precios_acciones[2] * acciones_portafolio["Amazon"]
valor_tesla = precios_acciones[3] * acciones_portafolio["Tesla"]
valor_google = precios_acciones[4] * acciones_portafolio["Google"]

print(valor_microsoft)
print(valor_amazon)
print(valor_tesla)
print(valor_google)

print((valor_tesla * 10/100) + valor_tesla)
```

## Parte 5 — Lista del portafolio

```python
valor_portafolio = []
valor_portafolio.append(valor_apple)
# Agrega los demás valores calculados
```
